```python
import numpy as np
arr = np.array([1,2,3,4,5])
result = arr **2    # Vectorization
result
```

```python
result + 10   # Broadcasting
```

```python
arr1 = np.array([[1,2,3], [4,5,6]])
arr2 = np.array([7,8,9])
result = arr1 + arr2  #Broadcasting arr1 across arr2
result
```

```python
# simulating the dataset (5 samples , 3 features)
data = np.array([[10,20,30],
                 [40,50,60],
                 [20,30,40],
                 [25,45,65],
                 [70,80,90]])
# calculating Mean and standard deviation for each feature (coloumn)
mean = data.mean(axis=0)
std = data.std(axis=0)

#Normalization the data using broadcasting
normalized_data = (data - mean) / std
print(normalized_data)
```
